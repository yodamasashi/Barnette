```python
import numpy as np
import networkx as nx
import matplotlib.pyplot as plt
from mpl_toolkits.mplot3d import Axes3D
from collections import defaultdict
from collections import deque
```

```python
# 両端に0を追加する関数
def R(a):
    A = [0]
    for i in range(len(a)):
        A.append(a[i])
    A.append(0)
    return A

# 境界線を越えるための関数
def L(a):
    A = []
    for i in range(len(a)):
        if i%2==0:
            A.append(a[i])
        if i%2==1:
            A.append(1-a[i])
    return A

# 面を埋める関数
def write(a,b,c):
    if (a+b+c)%2==0:
        return 1
    if (a+b+c)%2==1:
        return 0
    
# xy平面の塗り分け関数    
def xy(a):
    if a==0:
        return 2
    if a==1:
        return 4
    
# yz平面の塗り分け関数    
def yz(a):
    if a==0:
        return 4
    if a==1:
        return 6
    
# zx平面の塗り分け関数    
def zx(a):
    if a==0:
        return 6
    if a==1:
        return 2
    
# 赤だけ薄くする関数
def pr(a):
    if a==2:
        a=3
    return a

# 緑だけ薄くする関数
def pg(a):
    if a==4:
        a=5
    return a

# 青だけ薄くする関数
def pb(a):
    if a==6:
        a=7
    return a

# 間に0を挟む関数
def zero(a):
    zero = [0]
    for i in range(len(a)):
        zero.append(a[i])
        zero.append(0)
    return zero

# 数列の奇数番目を抜く関数
def remove_odd(a):
    re_odd=[]
    for i in range(len(a)):
        if i%2==1:
            re_odd.append(a[i])
    return re_odd

def generate_cube(x, y, z):

    rx = R(x)
    ry = R(y)
    rz = R(z)

    lx = L(rx)
    ly = L(ry)
    lz = L(rz)

    crx = zero(rx)
    cry = zero(ry)
    crz = zero(rz)

    clx = zero(lx)
    cly = zero(ly)
    clz = zero(lz)

    fxy_0 = np.zeros((len(rx),len(ly)))
    for i in range(len(rx)):
        fxy_0[i][0]=rx[i]
    for i in range(len(ly)):
        fxy_0[0][i]=ly[i]
    for i in range(1,len(rx)):
        for j in range(1,len(ly)):
            fxy_0[i][j]=write(fxy_0[i-1][j],fxy_0[i][j-1],fxy_0[i-1][j-1])

    fyz_0 = np.zeros((len(ry),len(lz)))
    for i in range(len(ry)):
        fyz_0[i][0]=ry[i]
    for i in range(len(lz)):
        fyz_0[0][i]=lz[i]
    for i in range(1,len(ry)):
        for j in range(1,len(lz)):
            fyz_0[i][j]=write(fyz_0[i-1][j],fyz_0[i][j-1],fyz_0[i-1][j-1])

    fzx_0 = np.zeros((len(rz),len(lx)))
    for i in range(len(rz)):
        fzx_0[i][0]=rz[i]
    for i in range(len(lx)):
        fzx_0[0][i]=lx[i]
    for i in range(1,len(rz)):
        for j in range(1,len(lx)):
            fzx_0[i][j]=write(fzx_0[i-1][j],fzx_0[i][j-1],fzx_0[i-1][j-1])

    ezx_x = []
    for i in range(len(rx)):
        ezx_x.append(fzx_0[len(rz)-1][i])
    urx = L(ezx_x)

    exy_x = []
    for i in range(len(lx)):
        exy_x.append(fxy_0[i][len(ry)-1])
    ulx = L(exy_x)

    exy_y = []
    for i in range(len(ry)):
        exy_y.append(fxy_0[len(rx)-1][i])
    ury = L(exy_y)

    eyz_y = []
    for i in range(len(ly)):
        eyz_y.append(fyz_0[i][len(rz)-1])
    uly = L(eyz_y)

    eyz_z = []
    for i in range(len(rz)):
        eyz_z.append(fyz_0[len(ry)-1][i])
    urz = L(eyz_z)

    ezx_z = []
    for i in range(len(lz)):
        ezx_z.append(fzx_0[i][len(rx)-1])
    ulz = L(ezx_z)

    fxy_1 = np.zeros((len(urx),len(uly)))
    for i in range(len(urx)):
        fxy_1[i][0]=urx[i]
    for i in range(len(uly)):
        fxy_1[0][i]=uly[i]
    for i in range(1,len(rx)):
        for j in range(1,len(ly)):
            fxy_1[i][j]=write(fxy_1[i-1][j],fxy_1[i][j-1],fxy_1[i-1][j-1])

    fyz_1 = np.zeros((len(ury),len(ulz)))
    for i in range(len(ury)):
        fyz_1[i][0]=ury[i]
    for i in range(len(ulz)):
        fyz_1[0][i]=ulz[i]
    for i in range(1,len(ury)):
        for j in range(1,len(ulz)):
            fyz_1[i][j]=write(fyz_1[i-1][j],fyz_1[i][j-1],fyz_1[i-1][j-1])

    fzx_1 = np.zeros((len(urz),len(ulx)))
    for i in range(len(urz)):
        fzx_1[i][0]=urz[i]
    for i in range(len(ulx)):
        fzx_1[0][i]=ulx[i]
    for i in range(1,len(urz)):
        for j in range(1,len(ulx)):
            fzx_1[i][j]=write(fzx_1[i-1][j],fzx_1[i][j-1],fzx_1[i-1][j-1])

    for i in range(len(rx)):
        for j in range(len(ly)):
            fxy_0[i][j]=xy(fxy_0[i][j])
            fxy_1[i][j]=xy(fxy_1[i][j])

    for i in range(len(ry)):
        for j in range(len(lz)):
            fyz_0[i][j]=yz(fyz_0[i][j])
            fyz_1[i][j]=yz(fyz_1[i][j])

    for i in range(len(rz)):
        for j in range(len(lx)):
            fzx_0[i][j]=zx(fzx_0[i][j])
            fzx_1[i][j]=zx(fzx_1[i][j])

    CUBE = np.zeros((2*len(x)+5,2*len(y)+5,2*len(z)+5))

    for i in range(len(rx)):
        for j in range(len(ly)):
            CUBE[2*i+1][2*j+1][0]=fxy_0[i][j]
            CUBE[2*i+1][2*j+1][2*len(rz)]=fxy_1[i][j]

    for i in range(len(ry)):
        for j in range(len(lz)):
            CUBE[0][2*i+1][2*j+1]=fyz_0[i][j]
            CUBE[2*len(rx)][2*i+1][2*j+1]=fyz_1[i][j]

    for i in range(len(rz)):
        for j in range(len(lx)):
            CUBE[2*j+1][0][2*i+1]=fzx_0[i][j]
            CUBE[2*j+1][2*len(ry)][2*i+1]=fzx_1[i][j]

    for i in range(2*len(rz)+1):
        if i%4==3:
            for j in range(2*len(rx)+1):
                for k in range(2*len(ry)+1):
                    CUBE[j][k][i]=pb(CUBE[j][k][i])

    for i in range(2*len(ry)+1):
        if i%4==3:
            for j in range(2*len(rz)+1):
                for k in range(2*len(rx)+1):
                    CUBE[k][i][j]=pg(CUBE[k][i][j])

    for i in range(2*len(rx)+1):
        if i%4==3:
            for j in range(2*len(ry)+1):
                for k in range(2*len(rz)+1):
                    CUBE[i][j][k]=pr(CUBE[i][j][k])

    for i in range(1,2*len(rx)):
        for j in range(2,2*len(ry)-1):
            if (i%2==1 and j%2==0):
                if CUBE[i][j-1][0]==CUBE[i][j+1][0]:
                    CUBE[i][j][0]=8
                else:
                    CUBE[i][j][0]=9
                if CUBE[i][j-1][2*len(rz)]==CUBE[i][j+1][2*len(rz)]:
                    CUBE[i][j][2*len(rz)]=8
                else:
                    CUBE[i][j][2*len(rz)]=9

    for i in range(2,2*len(rx)-1):
        for j in range(1,2*len(ry)):
            if (i%2==0 and j%2==1):
                if CUBE[i-1][j][0]==CUBE[i+1][j][0]:
                    CUBE[i][j][0]=8
                else:
                    CUBE[i][j][0]=9
                if CUBE[i-1][j][2*len(rz)]==CUBE[i+1][j][2*len(rz)]:
                    CUBE[i][j][2*len(rz)]=8
                else:
                    CUBE[i][j][2*len(rz)]=9

    for i in range(1,2*len(ry)):
        for j in range(2,2*len(rz)-1):
            if (i%2==1 and j%2==0):
                if CUBE[0][i][j-1]==CUBE[0][i][j+1]:
                    CUBE[0][i][j]=8
                else:
                    CUBE[0][i][j]=9
                if CUBE[2*len(rx)][i][j-1]==CUBE[2*len(rx)][i][j+1]:
                    CUBE[2*len(rx)][i][j]=8
                else:
                    CUBE[2*len(rx)][i][j]=9

    for i in range(2,2*len(ry)-1):
        for j in range(1,2*len(rz)):
            if (i%2==0 and j%2==1):
                if CUBE[0][i-1][j]==CUBE[0][i+1][j]:
                    CUBE[0][i][j]=8
                else:
                    CUBE[0][i][j]=9
                if CUBE[2*len(rx)][i-1][j]==CUBE[2*len(rx)][i+1][j]:
                    CUBE[2*len(rx)][i][j]=8
                else:
                    CUBE[2*len(rx)][i][j]=9

    for i in range(1,2*len(rx)):
        for j in range(2,2*len(rz)-1):
            if (i%2==1 and j%2==0):
                if CUBE[i][0][j-1]==CUBE[i][0][j+1]:
                    CUBE[i][0][j]=8
                else:
                    CUBE[i][0][j]=9
                if CUBE[i][2*len(ry)][j-1]==CUBE[i][2*len(ry)][j+1]:
                    CUBE[i][2*len(ry)][j]=8
                else:
                    CUBE[i][2*len(ry)][j]=9

    for i in range(2,2*len(rx)-1):
        for j in range(1,2*len(rz)):
            if (i%2==0 and j%2==1):
                if CUBE[i-1][0][j]==CUBE[i+1][0][j]:
                    CUBE[i][0][j]=8
                else:
                    CUBE[i][0][j]=9
                if CUBE[i-1][2*len(ry)][j]==CUBE[i+1][2*len(ry)][j]:
                    CUBE[i][2*len(ry)][j]=8
                else:
                    CUBE[i][2*len(ry)][j]=9

    for i in range(2*len(rx)+1):
        if i%2==1:
            if CUBE[i][0][1]==CUBE[i][1][0]:
                CUBE[i][0][0]=8
            else:
                CUBE[i][0][0]=9
            if CUBE[i][2*len(ry)][2*len(rz)-1]==CUBE[i][2*len(ry)-1][2*len(rz)]:
                CUBE[i][2*len(ry)][2*len(rz)]=8
            else:
                CUBE[i][2*len(ry)][2*len(rz)]=9
            if CUBE[i][2*len(ry)][1]==CUBE[i][2*len(ry)-1][0]:
                CUBE[i][2*len(ry)][0]=8
            else:
                CUBE[i][2*len(ry)][0]=9
            if CUBE[i][1][2*len(rz)]==CUBE[i][0][2*len(rz)-1]:
                CUBE[i][0][2*len(rz)]=8
            else:
                CUBE[i][0][2*len(rz)]=9

    for i in range(2*len(ry)+1):
        if i%2==1:
            if CUBE[0][i][1]==CUBE[1][i][0]:
                CUBE[0][i][0]=8
            else:
                CUBE[0][i][0]=9
            if CUBE[2*len(rx)][i][2*len(rz)-1]==CUBE[2*len(rx)-1][i][2*len(rz)]:
                CUBE[2*len(rx)][i][2*len(rz)]=8
            else:
                CUBE[2*len(rx)][i][2*len(rz)]=9
            if CUBE[2*len(rx)][i][1]==CUBE[2*len(rx)-1][i][0]:
                CUBE[2*len(rx)][i][0]=8
            else:
                CUBE[2*len(rx)][i][0]=9
            if CUBE[1][i][2*len(rz)]==CUBE[0][i][2*len(rz)-1]:
                CUBE[0][i][2*len(rz)]=8
            else:
                CUBE[0][i][2*len(rz)]=9

    for i in range(2*len(rz)+1):
        if i%2==1:
            if CUBE[0][1][i]==CUBE[1][0][i]:
                CUBE[0][0][i]=8
            else:
                CUBE[0][0][i]=9
            if CUBE[2*len(rx)][2*len(ry)-1][i]==CUBE[2*len(rx)-1][2*len(ry)][i]:
                CUBE[2*len(rx)][2*len(ry)][i]=8
            else:
                CUBE[2*len(rx)][2*len(ry)][i]=9
            if CUBE[2*len(rx)][1][i]==CUBE[2*len(rx)-1][0][i]:
                CUBE[2*len(rx)][0][i]=8
            else:
                CUBE[2*len(rx)][0][i]=9
            if CUBE[1][2*len(ry)][i]==CUBE[0][2*len(ry)-1][i]:
                CUBE[0][2*len(ry)][i]=8
            else:
                CUBE[0][2*len(ry)][i]=9

    return CUBE

def create_graph(x, y, z, CUBE):
    G = nx.grid_graph((len(z)+3, len(y)+3, len(x)+3))
    for i in range(1, len(x)+2):
        for j in range(1, len(y)+2):
            for k in range(1, len(z)+2):
                G.remove_node((i, j, k))

    edges = list(G.edges)
    for i in range(len(edges)):
        if CUBE[edges[i][0][0]+edges[i][1][0]][edges[i][0][1]+edges[i][1][1]][edges[i][0][2]+edges[i][1][2]]==8:
            if G.has_edge(edges[i][0], edges[i][1]):
                G.remove_edge(edges[i][0], edges[i][1])
    
    return G

def planar_dual_from_embedding(G):
    """
    平面グラフ G の平面埋め込みから双対グラフを作る。
    戻り値: dual（極大平面的グラフ）
    """
    # 平面性のチェックと平面埋め込みの取得
    is_planar, embedding = nx.check_planarity(G)
    if not is_planar:
        raise nx.NetworkXException("G が平面グラフではありません")

    # 各半辺 (u, v) がどの面に属するかを記録する
    halfedge_face = {}
    used = set()
    faces = []

    for u in embedding:
        for v in embedding.neighbors_cw_order(u):
            if (u, v) in used:
                continue

            # (u, v) の右側にある 1 つの面をたどる
            mark = set()
            face_nodes = embedding.traverse_face(u, v, mark)
            fid = len(faces)
            faces.append(face_nodes)

            # この面に属する全ての半辺に face ID を振る
            for he in mark:
                halfedge_face[he] = fid
            used |= mark

    # 双対グラフ：頂点 = 面，辺 = 面同士が共有する元の辺
    dual = nx.Graph()
    dual.add_nodes_from(range(len(faces)))  # 0,1,2,...,F-1

    for u, v in G.edges():
        f1 = halfedge_face.get((u, v))
        f2 = halfedge_face.get((v, u))
        # どちらか face が取れない / 同じ face の場合はスキップ
        if f1 is None or f2 is None or f1 == f2:
            continue
        dual.add_edge(f1, f2)

    # おまけ：各双対頂点がどの面に対応するかを覚えておく
    dual.graph["faces"] = faces
    return dual

def generate_soutui_graph(x, y, z, CUBE):
    """
    現在の CUBE から作った 3 正則 3 連結平面 2 部グラフ G の
    双対グラフ（極大平面グラフ）を返す。
    CUBE 自体は create_graph() の中で既に使われているので，
    ここでは特に参照しなくても良い。
    """
    G_primal = create_graph(x, y, z, CUBE)               # あなたの既存の関数
    G_dual   = planar_dual_from_embedding(G_primal)
    return G_dual

def three_color_degree_stats_fast(G):
    """
    偶数次数の極大平面グラフ G（= Gs を想定）を 3-頂点彩色し，
    各色ごとに「次数 d の頂点が何個あるか」を返す。

    戻り値:
        colors : dict
            頂点 -> 色 (0, 1, 2)
        stats : dict
            { 色 : { 次数 : 個数 } }
    """

    # --- 1. グラフ中の三角形（3-サイクル）を全部拾う ---
    triangles_set = set()
    for u in G.nodes():
        nbrs = list(G.neighbors(u))
        ln = len(nbrs)
        for i in range(ln):
            v = nbrs[i]
            for j in range(i + 1, ln):
                w = nbrs[j]
                if G.has_edge(v, w):
                    tri = tuple(sorted((u, v, w)))
                    triangles_set.add(tri)

    triangles = list(triangles_set)
    if not triangles:
        raise ValueError("三角形が見つかりません。極大平面グラフではなさそうです。")

    # --- 2. 最初の三角形を 3 色で塗る（0,1,2 はどの順でもOK） ---
    colors = {v: None for v in G.nodes()}
    a, b, c = triangles[0]
    colors[a], colors[b], colors[c] = 0, 1, 2

    # --- 3. 「三角形のうち2頂点が塗れていたら残り1頂点を強制」で伝播 ---
    changed = True
    ALL_COLORS = {0, 1, 2}

    while changed:
        changed = False
        for u, v, w in triangles:
            cu, cv, cw = colors[u], colors[v], colors[w]

            # 3頂点すべて塗れている場合は，三角形が3色になっているかチェック
            if cu is not None and cv is not None and cw is not None:
                if len({cu, cv, cw}) != 3:
                    raise ValueError("矛盾: 三角形が 3 色になっていません。")
                continue

            # ちょうど2頂点だけ色が決まっている場合，残り1頂点の色が強制される
            if cu is None and cv is not None and cw is not None:
                used = {cv, cw}
                avail = ALL_COLORS - used
                if len(avail) != 1:
                    raise ValueError("矛盾: 頂点の色が一意に定まりません。")
                colors[u] = avail.pop()
                changed = True

            elif cv is None and cu is not None and cw is not None:
                used = {cu, cw}
                avail = ALL_COLORS - used
                if len(avail) != 1:
                    raise ValueError("矛盾: 頂点の色が一意に定まりません。")
                colors[v] = avail.pop()
                changed = True

            elif cw is None and cu is not None and cv is not None:
                used = {cu, cv}
                avail = ALL_COLORS - used
                if len(avail) != 1:
                    raise ValueError("矛盾: 頂点の色が一意に定まりません。")
                colors[w] = avail.pop()
                changed = True

    # --- 4. すべての頂点に色が付いているか確認 ---
    uncolored = [v for v, c in colors.items() if c is None]
    if uncolored:
        raise ValueError(
            f"彩色されていない頂点が残りました: {uncolored[:10]} ..."
            "（グラフが偶数次数の極大平面グラフでない可能性があります）"
        )

    # --- 5. 色ごとの次数分布を集計 ---
    stats = defaultdict(lambda: defaultdict(int))
    for v, col in colors.items():
        d = G.degree(v)
        stats[col][d] += 1

    return colors, stats

# ここから下を追加してください

class UnionFind:
    """森（サイクルができないように）を管理するための Union-Find"""
    def __init__(self, nodes=()):
        self.parent = {v: v for v in nodes}
        self.rank   = {v: 0 for v in nodes}

    def add(self, x):
        if x not in self.parent:
            self.parent[x] = x
            self.rank[x] = 0

    def find(self, x):
        # 経路圧縮付き find
        while self.parent[x] != x:
            self.parent[x] = self.parent[self.parent[x]]
            x = self.parent[x]
        return x

    def union(self, x, y):
        """x と y を結ぶ辺を追加する。
           すでに同一連結成分なら False（＝サイクルができる）を返す。
        """
        rx, ry = self.find(x), self.find(y)
        if rx == ry:
            return False
        if self.rank[rx] < self.rank[ry]:
            rx, ry = ry, rx
        self.parent[ry] = rx
        if self.rank[rx] == self.rank[ry]:
            self.rank[rx] += 1
        return True

    def clone(self):
        """バックトラック用に軽くコピー"""
        new = UnionFind()
        new.parent = self.parent.copy()
        new.rank   = self.rank.copy()
        return new


def two_forest_recolor_search(G, colors, degree_threshold=6):
    """
    問題文の処理を実装した関数。
    （偶数次数の極大平面グラフ G ＋ 3-頂点彩色 colors を仮定）

    - 途中で「2つの誘導部分グラフが両方ともサイクルを含まない」
      振り分けパターンが見つかれば 0 を返す。
    - どの集合を基準にしてもダメなら 1 を返す。
    """
    # 次数を先に全部取っておく
    deg = dict(G.degree())

    # 色ごとの頂点集合
    color_sets = {c: set() for c in range(3)}
    for v, col in colors.items():
        color_sets[col].add(v)

    # 各色ごとの「次数 >= degree_threshold の頂点」リスト
    high_nodes = {
        c: [v for v in color_sets[c] if deg[v] >= degree_threshold]
        for c in range(3)
    }

    # 「高次数頂点の個数が少ない集合」から順に基準色として試す
    order = sorted(range(3), key=lambda c: len(high_nodes[c]))

    def build_union_find(nodes):
        """
        与えられた頂点集合 nodes 上の誘導部分グラフが森かどうかを確認しつつ，
        その森を表す UnionFind を返す。
        サイクルがあれば None を返す。
        """
        H = G.subgraph(nodes)
        uf = UnionFind(H.nodes())
        for u, v in H.edges():
            if not uf.union(u, v):
                # サイクルが発生
                return None
        return uf

    def try_base_color(base_color):
        """
        ある基準色 base_color を固定して，
        その色に属する次数 >= 6 の頂点たちを，
        残り 2 色に振り分けていけるかを探索する。
        """
        base_high = high_nodes[base_color]
        # 残り 2 色
        others = [c for c in range(3) if c != base_color]
        a, b = others[0], others[1]

        # 最初の set_a, set_b
        S_a = set(color_sets[a])
        S_b = set(color_sets[b])

        # 初期状態でそれぞれ森かどうかチェック（ダメならこの基準色は不可能）
        uf_a = build_union_find(S_a)
        uf_b = build_union_find(S_b)
        if uf_a is None or uf_b is None:
            return False

        # 移動させる高次数頂点がそもそも無ければ，
        # この時点で「2つとも森」なので成功
        if not base_high:
            return True

        # 高次数頂点は次数が大きい順に処理すると枝刈りしやすい
        H_sorted = sorted(base_high, key=lambda v: -deg[v])
        n_high = len(H_sorted)

        def backtrack(i, S_a_curr, S_b_curr, uf_a_curr, uf_b_curr):
            """
            i 番目までの高次数頂点を振り分けた状態から，
            さらに先へ振り分けていくバックトラック。
            途中でどこかサイクルが出た枝は即座に打ち切る。
            """
            if i == n_high:
                # 全て振り分け終わってもサイクルなし → 成功
                return True

            v = H_sorted[i]

            # --- ケース1: v を色 a 側に移す ---
            S_a_next = S_a_curr | {v}
            uf_a_next = uf_a_curr.clone()
            uf_a_next.add(v)
            ok_a = True
            for u in G.neighbors(v):
                if u in S_a_curr:   # すでに色 a 側にいる頂点との辺だけ見る
                    if not uf_a_next.union(v, u):
                        ok_a = False
                        break
            if ok_a:
                if backtrack(i + 1, S_a_next, S_b_curr, uf_a_next, uf_b_curr):
                    return True

            # --- ケース2: v を色 b 側に移す ---
            S_b_next = S_b_curr | {v}
            uf_b_next = uf_b_curr.clone()
            uf_b_next.add(v)
            ok_b = True
            for u in G.neighbors(v):
                if u in S_b_curr:
                    if not uf_b_next.union(v, u):
                        ok_b = False
                        break
            if ok_b:
                if backtrack(i + 1, S_a_curr, S_b_next, uf_a_curr, uf_b_next):
                    return True

            # どちらに入れてもサイクルが出来てしまう
            return False

        return backtrack(0, S_a, S_b, uf_a, uf_b)

    # 高次数頂点の少ない色から順に試す
    for base in order:
        if try_base_color(base):
            return 0  # 条件を満たす振り分けが見つかった

    # どの色を基準にしてもダメだった
    return 1

def to_binary1(n):
    """整数 n を 2 進数表現のリストに変換する関数"""
    if n == 0:
        return [0]
    bits = []
    while n > 0:
        bits.append(n % 2)
        n //= 2
    while len(bits) < 1:
        bits.append(0)
    return bits

def to_binary3(n):
    """整数 n を 2 進数表現のリストに変換する関数"""
    if n == 0:
        return [0,0,0]
    bits = []
    while n > 0:
        bits.append(n % 2)
        n //= 2
    while len(bits) < 3:
        bits.append(0)
    return bits

def to_binary5(n):
    """整数 n を 2 進数表現のリストに変換する関数"""
    if n == 0:
        return [0,0,0,0,0]
    bits = []
    while n > 0:
        bits.append(n % 2)
        n //= 2
    while len(bits) < 5:
        bits.append(0)
    return bits

def to_binary7(n):
    """整数 n を 2 進数表現のリストに変換する関数"""
    if n == 0:
        return [0,0,0,0,0,0,0]
    bits = []
    while n > 0:
        bits.append(n % 2)
        n //= 2
    while len(bits) < 7:
        bits.append(0)
    return bits

def to_binary9(n):
    """整数 n を 2 進数表現のリストに変換する関数"""
    if n == 0:
        return [0,0,0,0,0,0,0,0,0]
    bits = []
    while n > 0:
        bits.append(n % 2)
        n //= 2
    while len(bits) < 9:
        bits.append(0)
    return bits

import networkx as nx

def has_two_or_more_components(G: nx.Graph) -> bool:
    """
    G: networkx.Graph（無向グラフ）を想定
    連結成分が 2 個以上あるなら True、それ以外なら False を返す
    """
    # ノードが 0 個なら連結成分は 0 個とみなして False
    if G.number_of_nodes() == 0:
        return False

    return nx.number_connected_components(G) >= 2
    # あるいは: return not nx.is_connected(G)

```

```python
from typing import Iterable, Tuple, List, Optional, Sequence, Set

Edge = Tuple[int, int]

def find_hamiltonian_cycle_from_edges(
    edges: Iterable[Edge],
    vertices: Optional[Sequence[int]] = None,
    directed: bool = False,
) -> Optional[List[int]]:
    """
    edges: 辺リスト（各辺は (u, v) のタプル。u, v は任意の整数）
    vertices: （任意）頂点集合を明示したい場合に渡す（孤立頂点を含めたい等）
              省略時は edges に登場する頂点全体を頂点集合とする
    directed: False=無向, True=有向（有向なら (u, v) は u->v）

    戻り値:
      ハミルトン閉路が見つかれば、頂点ラベルの列 [v0, v1, ..., v(n-1), v0] を返す
      見つからなければ None
    """
    edges = list(edges)

    # 頂点集合を確定
    vs: Set[int] = set(vertices) if vertices is not None else set()
    self_loops: Set[int] = set()
    for u, v in edges:
        vs.add(u)
        vs.add(v)
        if u == v:
            self_loops.add(u)

    if not vs:
        return None

    # 頂点ラベル -> 0..n-1 に圧縮
    labels = sorted(vs)
    idx = {v: i for i, v in enumerate(labels)}
    n = len(labels)

    # n==1 の特別扱い（自己ループがあれば閉路）
    if n == 1:
        v = labels[0]
        return [v, v] if v in self_loops else None

    # 隣接（集合で保持して membership を O(1) に）
    adj: List[Set[int]] = [set() for _ in range(n)]
    for u, v in edges:
        iu, iv = idx[u], idx[v]
        if iu == iv:
            continue  # 自己ループはここでは探索に使わない（n==1は上で処理済み）
        adj[iu].add(iv)
        if not directed:
            adj[iv].add(iu)

    # 必要条件の軽い枝刈り（n>2 のとき）
    if not directed:
        if n > 2 and any(len(adj[i]) < 2 for i in range(n)):
            return None
    else:
        if n > 1:
            indeg = [0] * n
            outdeg = [0] * n
            for i in range(n):
                outdeg[i] = len(adj[i])
                for j in adj[i]:
                    indeg[j] += 1
            if any(indeg[i] == 0 or outdeg[i] == 0 for i in range(n)):
                return None

    # 開始点はどこでも良いので、分岐が少なそうな点から始める（軽いヒューリスティック）
    start = min(range(n), key=lambda i: len(adj[i]))

    visited = [False] * n
    visited[start] = True
    path = [start]

    # 近傍の順序も少し工夫（次数が小さい順）
    neigh_order = [sorted(list(adj[i]), key=lambda j: len(adj[j])) for i in range(n)]

    def dfs(v: int, depth: int) -> bool:
        if depth == n:
            return start in adj[v]
        for to in neigh_order[v]:
            if not visited[to]:
                visited[to] = True
                path.append(to)
                if dfs(to, depth + 1):
                    return True
                path.pop()
                visited[to] = False
        return False

    if not dfs(start, 1):
        return None

    path.append(start)
    return [labels[i] for i in path]


def has_hamiltonian_cycle_from_edges(
    edges: Iterable[Edge],
    vertices: Optional[Sequence[int]] = None,
    directed: bool = False,
) -> bool:
    return find_hamiltonian_cycle_from_edges(edges, vertices, directed) is not None


# --- 使い方例 ---

```
